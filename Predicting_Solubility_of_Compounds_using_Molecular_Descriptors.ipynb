# **First ML building Model Project**

## **Load Data**

```python
import pandas as pd
```

```python
Data = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
```

```python
Data
```

## **Data Preparation**

### **Data Separation as X and Y**

```python
y = Data['logS']
y
```

```python
x = Data.drop('logS', axis = 1)
x
```

### **Data Spliting**

```python
print(x.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
(x_train)
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## **Model Building**

### **Linear Regression**

```python
from sklearn.linear_model import LinearRegression
```

#### **Training the model**

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

#### **Applying the model to make a prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### **Evaluate the model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)
```

```python
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_result = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_result.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_result
```

## **Random Forest**

#### **Training the model**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

#### **Applying the model to make a prediction**

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

### **Evaluation model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)
```

```python
rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
print('RF MSE (Train): ', rf_train_mse)
print('RF R2 (Train): ', rf_train_r2)
print('RF MSE (Test): ', rf_test_mse)
print('RF R2 (Test): ', rf_test_r2)
```

```python
rf_result = pd.DataFrame(['Linear regression', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_result.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
rf_result
```

## **Model Comparison**

```python
df_model = pd.concat([lr_result, rf_result], axis=0)
df_model.index = range(df_model.shape[0])
df_model
```
